````python

#Total Population distribtion by Different regions of world

import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('POPULATION.csv')





region_size = 5
df['Region'] = df.index // region_size


region_population = df.groupby('Region')['Total Population'].sum()


plt.figure(figsize=(40, 20))
plt.bar(region_population.index, region_population, color="skyblue", edgecolor="black")


plt.xlabel("Region (Groups of 5 countries)")
plt.ylabel("Total Population (Billions)")
plt.title("Population by Region in 2023")
plt.xticks(rotation=45)


plt.show()


````

````python

#Population by Gender differentiation....

import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('POPULATION.csv')





region_size = 5
df['Region'] = df.index // region_size


region_population = df.groupby('Region')[['Total Population', 'Male Population', 'Female Population']].sum()


region_population.plot(kind='bar', figsize=(10, 6), color=["skyblue", "lightcoral", "lightgreen"], edgecolor="black")


plt.xlabel("Region (Groups of 5 countries)")
plt.ylabel("Population (Billions)")
plt.title("Population by Region in 2023: Total, Male, and Female")
plt.xticks(rotation=45)


plt.show()

````
